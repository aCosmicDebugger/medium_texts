# Introdução:

A **regressão linear** é um método estatístico utilizado para estabelecer uma relação, no caso linear, entre uma **variável resposta** $y$ e uma ou mais **variáveis explicativas** $x$. Na prática, a situação é a seguinte: existem pares de valores ($x, y$) que são medidos, como por exemplo, o tamanho da cauda do animal e seu tamanho, pressão do sangue e peso, ou a taxa de poupança interna e o crescimento econômico.

Esses pares de valores são representados por pontos em um plano cartesiano, e o conjunto desses pontos (nosso conjunto de dados) pode indicar alguma relação linear. Caso isso seja verdade, procuramos encontrar a reta que melhor se ajusta a esse conjunto de dados. Para fazer isso, definimos uma reta da forma:

$$
\hat{y}_{i} = a x_{i} + b \,\,\,\, ,
$$

com $i = 1, \ldots, N$, onde $N$ é o número de observações. Desejamos que os valores $\hat{y}$ da reta se aproximem ao máximo dos valores reais $y$. Para alcançar isso, é necessário garantir a existência de valores $a$ e $b$ que proporcionem esse ajuste. Podemos entender como isso ocorre através do método dos mínimos quadrados ordinários.

Em resumo, o que estamos fazendo é escolher uma função preditora cujos coeficientes regressores (parâmetros desconhecidos) do modelo são obtidos a partir do conjunto de dados. No exemplo acima, o modelo é de apenas uma variável, mas se estivéssemos lidando com mais de uma variável explicativa, chamaríamos isso de uma regressão de múltiplas variáveis, e o raciocínio seria análogo.

# OLSM - Método dos Mínimos Quadrados Ordinários:

O método dos mínimos quadrados ordinários é uma técnica utilizada para ajustar uma linha ou curva a um conjunto de dados. O objetivo é encontrar os coeficientes que minimizam a soma dos quadrados dos resíduos, ou seja, as diferenças entre os valores observados e os valores preditos pelo modelo. Este método é comumente aplicado em regressões lineares para encontrar a melhor aproximação linear aos dados.

A diferença entre o $i$-ésimo valor da variável do modelo $\hat{y}_{i}$ e o $i$-ésimo valor de $y$ é chamada de **resíduo**, denotado por $r$:

$$
r_{i} = \hat{y}_{i} - y_{i} \,\,\, .
$$

Vamos escrever o resíduo então como:

$$
r_{i} = \hat{y}_{i} - \left(ax_{i}-b\right) \,\,\, .
$$

Aqui, deve-se notar que, como definido acima, o resíduo $r$ pode ser maior ou menor que zero. O que é de nosso interesse não é apenas um único $r_{i}$ específico, mas sim o resíduo total, isto é, a soma dos resíduos. Neste ponto encontramos o seguinte problema: este resíduo tanto pode ser positivo quanto negativo, assim é possível (porém não exatamente provável) que a sua soma seja igual a zero! O que não faria muito sentido. Portanto, além do resíduo total, estamos interessados no _valor absoluto_ dos resíduos. Sendo assim, a quantidade de interesse é:

$$
S = \sum_{i=1}^{N} r^{2}_{i} = \sum_{i=1}^{N} \left( \hat{y}_{i} - ax_{i} - b \right)^{2} \,\,\, .
$$

Chamamos $S$ de _função objetivo_; desejamos **minimizar** $S$ em relação às incógnitas ($a, b$). Derivamos a expressão acima em relação às incógnitas e igualamos a zero. Então, derivando temos:

$$
\frac{\partial S}{\partial a} =  -2 \sum_{i=1}^{N} \left( \hat{y}_{i} - ax_{i} - b\right)x_{i} = 0 \,\,\, , 
$$

e

$$
\frac{\partial S}{\partial b} = -2 \sum_{i=1}^{N} \left( \hat{y}_{i} - ax_{i} - b\right) = 0 \,\,\,.
$$

Percebemos aqui que conseguimos duas equações com duas incógnitas. Sendo elas, 

$$
\sum_{i=1}^{N} \hat{y}_{i} x_{i} = a \sum_{i=1}^{N}  x_{i}^{2} +   b\sum_{i=1}^{N}  x_{i}
$$

e

$$
\sum_{i=1}^{N}\hat{y}_{i} =  a   \sum_{i=1}^{N} x_{i} + b N \,\,\,\, .
$$

Para resolver esse sistema de equações, vamos multiplicar a primeira por $\sum_{i=1}^{N} x_{i}$, a segunda por $\sum_{i=1}^{N} x^{2}_{i}$ e subtrair uma da outra, assim encontrando o valor de $b$:

$$
\sum_{i}^{N} \hat{y}_{ì} \,x_{i} \cdot \sum_{i}^{N}x_{i} = a \sum_{i}^{N}x^{2}_{i}  \cdot \sum_{i}^{N}x_{i} + b  \sum_{i}^{N}x_{i} \cdot  \sum_{i}^{N}x_{i}
$$

$$
\sum_{i}^{N} \hat{y}_{i}  \cdot \sum_{i}^{N}x^{2}_{i} = a \sum_{i}^{N}x_{i}\cdot  \sum_{i}^{N}x^{2}_{i} + bN  \sum_{i}^{N}x^{2}_{i}
$$

Ao fazer isso, chegamos à expressão:

$$
\sum_{i}^{N} \hat{y}_{ì} \,x_{i} \cdot \sum_{i}^{N}x_{i}  - \sum_{i}^{N} \hat{y}_{i}  \cdot \sum_{i}^{N}x^{2}_{i} = b \left[   \sum_{i}^{N}x_{i} \cdot  \sum_{i}^{N}x_{i} -  N  \sum_{i}^{N}x^{2}_{i} \right]
$$

de onde isolamos $b$ para encontrar:

$$
b = \frac{\sum_{i}^{N} \hat{y}_{ì} \,x_{i} \cdot \sum_{i}^{N}x_{i}  - \sum_{i}^{N} \hat{y}_{i}  \cdot \sum_{i}^{N}x^{2}_{i} }{\sum_{i}^{N}x_{i} \cdot  \sum_{i}^{N}x_{i} -  N  \sum_{i}^{N}x^{2}_{i}}
$$

Agora ao multiplicarmos a primeira por $N$ e a segunda por $\sum_{i=1}^{N} x_{i}$:

$$
N \sum_{i=1}^{N} \hat{y}_{i} x_{i} = a\,N \sum_{i=1}^{N}  x_{i}^{2} +   b\,N \sum_{i=1}^{N}  x_{i}
$$

$$
\sum_{i=1}^{N}\hat{y}_{i}\sum_{i=1}^{N} x_{i} =  a   \sum_{i=1}^{N} x_{i} \sum_{i=1}^{N} x_{i} + b N\sum_{i=1}^{N} x_{i} \,\,\,\, .
$$

Agora subtraindo a segunda da primeira, chegamos a

$$
 N \sum_{i=1}^{N} \hat{y}_{i} x_{i} - \sum_{i=1}^{N}\hat{y}_{i}\sum_{i=1}^{N} x_{i} = a \left[ N \sum_{i=1}^{N}  x_{i}^{2} - \sum_{i=1}^{N} x_{i} \sum_{i=1}^{N} x_{i}  \right]\,\,\,\, , 
$$

expressão que nos diz que a é dado por:

$$
a = \frac{N \sum_{i=1}^{N} \hat{y}_{i} x_{i} - \sum_{i=1}^{N}\hat{y}_{i}\sum_{i=1}^{N} x_{i} }{N \sum_{i=1}^{N}  x_{i}^{2} - \sum_{i=1}^{N} x_{i} \sum_{i=1}^{N} x_{i} }\,\,\,\, . 
$$

Para checar se realmente são mínimos, basta calcular as segundas derivadas, $\frac{\partial^{2} S}{\partial a^{2}}$ e $\frac{\partial^{2} S}{\partial b^{2}}$, além das derivadas mistas, $\frac{\partial^{2} S}{\partial a \partial b}$ e $\frac{\partial^{2} S}{\partial a \partial b}$ escrever a matriz  *Hessiana*  e checarse o determinante é realmente maior que zero.

$$
\frac{\partial^2 S}{\partial a^2} = 2 \sum_{i=1}^{N} x_{i}^{2} 
$$

$$
\frac{\partial^2 S}{\partial b^2} = 2N 
$$

$$
\frac{\partial^2 S}{\partial a\,  \partial b} = \frac{\partial^2 S}{\partial b\, \partial a} = 2\sum_{i}^{N} x_{i}
$$

A matriz $A$, das derivadas, é:

$$
A = \begin{bmatrix}
2\sum_{i=1}^{N} x_{i}^{2} & 2\sum_{i=1}^{N} x_{i} \\
2\sum_{i=1}^{N} x_{i} & 2N
\end{bmatrix}
$$

cujo determinante será dado por:

$$
\text{det} \left(A\right) = 4\left( N\sum_{i=1}^{N} x_{i}^{2}  - \sum_{i=1}^{N} x_{i} \cdot \sum_{i=1}^{N} x_{i}\right)  > 0
$$

Que, podemos ver, é maior que zero pois o produto $\sum_{i}^{N}x_{i}$ por ele mesmo será uma quantidade positiva e $x^{2}$ é necessariamente positivo. Restaria mostrar que a soma do segundo termo será sempre menor que soma do primeiro. Usando a desigualdade de Cauchy-Schwarz, onde para quaisquer sequências de números reais $a_1, a_2, \ldots, a_n$ e $(b_1, b_2, \ldots, b_n$, a seguinte inequação é verdadeira:

$$
\left(\sum_{i=1}^{n} a_i²\right) \left(\sum_{i=1}^{n} b_i²\right) \geq \left(\sum_{i=1}^{n} a_ib_i\right)²
$$

Se aplicarmos isso à nossa expressão, considerando $a_i = x_i$ e $b_i = 1$, obtemos:

$$
\left(\sum_{i=1}^{N} x^{2}_{i}\right) \left(\sum_{i=1}^{N} 1²\right) \geq \left(\sum_{i=1}^{N} x_i\right)²
$$

Simplificando, temos:

$$
\left(\sum_{i=1}^{N} x^{2}_{i}\right) \geq \left(\sum_{i=1}^{N} x_i\right)²
$$  

Portanto, a expressão $\sum_{i=1}^{N} x^{2}_{i} - \left(\sum_{i=1}^{N} x_{i}\right)^{2}$ é sempre maior ou igual a zero.  Nossos coeficientes estão completamente determinados e podemos utilizá-los na nossa equação do modelo, ou seja, a equação para $\hat{y}$.

>Existem outros métodos que podemos utilizar para verificar quais os parâmetros da nossa curva de _melhor ajuste_, mas o método de mínimos quadráticos tem a seguinte vantagem: a nossa função custo ou objetivo é **simétrica**. Desta maneira, um erro abaixo da curva ou acima da curva, desde que ambos de mesma magnitude, representam a mesma perda.

Por fim,  vamos aproveitar e  reescrever toda a expressão para os coeficientes usando: $\bar{x} = \frac{1}{N} \sum_{i=1}^{N} x_{i}$. Vamos multiplicar e dividir ambas as equações de $a$ e de $b$ por $N^{2}$, assim obtemos:

$$
a = \frac{\bar{yx }- \bar{y}\bar{x}}{\bar{x^{2}}- \bar{x}^{2}}
$$

$$
b = \frac{\bar{x^{2} } \bar{y} - \bar{x}\bar{yx} }{\bar{x^{2}}- \bar{x}^{2}}
$$

Lembrando que invertemos a subtração do denominador de $b$ bem como o numerador. Interessante de ver que nossos coeficientes acabam não dependendo do tamanho da amostra ($N$).


# Coeficiente de determinação, R$^{2}$:

Para concluir, existe uma medida do poder explicativo do nosso modelo, o R$^{2}$ , conhecido como **coeficiente de determinação**, não deve ser confundido com o coeficiente de correlação R de Pearson. Queremos saber como poderíamos medir este poder preditivo do nosso modelo? Voltando para a equação da soma dos quadrados dos resíduos, temos:

$$
S = \sum_{i}^{N} r^{2}_{i} = \sum_{i}^{N} \left( y_{i} - \hat{y}_{i} \right)^{2}
$$

Nós devemos interpretar essa função $S$ como o quanto nossa previsão difere do valor real da medida de $y$.
Isto é basicamente, em números, o quanto a nossa previsão difere dos valores reais. Agora, poderíamos
também definir as seguintes somas:

$$
S_{T} = \sum_{i}^{N} \left( y_{i} - \bar{y} \right)^{2}
$$

A diferença quadrática entre cada observação e a média das observações que é proporcional a variância
das observações, $\sigma^{2} = \frac{1}{N}\sum_{i=1}^{N} \left(y_{i} - \mu\right)^{2}$, sendo que $\mu$ é a média da amostra. E

$$
S_{E} = \sum_{i}^{N} \left( \hat{y} - \bar{y} \right)^{2}
$$

Que é a diferença entre a nossa estimativa e o valor médio das observações. Chamada de 'Soma explicada
dos quadrados' Podemos ver que:

$$
S_{T} - S = S_{E}
$$

A fração do que a nosso modelo não explica, seria a razão entre a variância dos erros, isto é a soma dos
quadrados dos resíduos S , sobre a variância total $S_{T}$:

$$
\frac{S}{S_{T}} = \frac{ S_{T} - S_{E}}{S_{T}} = 1 - \frac{S_{E}}{S_{T}}
$$

Finalmente, definimos o R$^{2}$ que é o quanto nosso modelo consegue explicar o que é proposto.

$$
\text{R}^{2} = 1 - \frac{S_{E}}{S_{T}}
$$

Em outras palavras, R$^{2}$  mede quão bom é o ajuste da curva, onde um valor igual 1 para o coeficiente implicaria em um ajuste perfeito.

# A descoberta da expansão do universo:

Na primeira metade do século XX, a compreensão do cosmos estava mergulhada na ideia de que a Via Láctea era a única galáxia existente, e as nebulosas eram consideradas objetos dentro de nossa própria galáxia. As nebulosas eram, até então, entendidas como nuvens de gás e poeira interestelar.  No entanto, Hubble contribuiu significativamente para redefinir essa visão. Utilizando o Telescópio Hooker no Observatório Mount Wilson, Hubble observou uma classe especial de nebulosas, as "nebulosas espirais", identificando-as como galáxias independentes distantes da Via Láctea. 

Hubble utilizou técnicas de observação cuidadosas para estudar essas nebulosas espirais. Ao analisar as imagens obtidas pelo telescópio, ele foi capaz de identificar estrelas individuais dentro dessas nebulosas. Isso foi crucial, pois permitiu a aplicação da técnica das variáveis Cefeidas. As estrelas Cefeidas são estrelas cujo brilho varia periodicamente ao longo do tempo. A relação entre o período dessa variação e a luminosidade intrínseca da estrela permitiu a Hubble estimar a distância até essas estrelas. Ao medir as distâncias das Cefeidas dentro das nebulosas espirais, Hubble percebeu que essas galáxias espirais estavam muito além dos limites da Via Láctea. 

Assim, ao identificar e medir as distâncias para as estrelas Cefeidas dentro dessas nebulosas espirais, Hubble foi capaz de demonstrar que elas eram, na verdade, galáxias independentes distantes da nossa própria galáxia. Além disso, no seu artigo: "[A relation between distance and radial velocity among extra-galactic nebulae](https://ui.adsabs.harvard.edu/abs/1929PNAS...15..168H/abstract)",  utilizou essas estrelas para calcular as distâncias até as galáxias, e então, combinando essas medidas com a velocidade de recessão observada nas linhas espectrais, formulou o que conhecemos hoje como **Lei de Hubble**. Ela é relação linear entre a distância de uma galáxia $\mathbf{r}$ e sua velocidade de afastamento $v$,

$$
v = H_{0} \mathbf{r} \,\,\,\, ,
$$

sendo $H_{0}$ a  chamada constante de Hubble, que indica a existência de uma expansão do universo. A importância dessas medições residia na confirmação de que o universo não apenas continha múltiplas galáxias, mas também estava em constante expansão, uma descoberta fundamental que impactou profundamente a cosmologia e nossa compreensão do cosmos.

A distância própria $\mathbf{r}$ pode ser obtida a partir das magnitudes aparente $m$ e absoluta $M$ usando a fórmula da relação de módulo de distância. A fórmula é a seguinte:

$$

$$
\mathbf{r} = 10^{ (m-M)/5 \,\,+ 1 }
$$

A distância própria $\mathbf{r}$ será dada em unidades arbitrárias, geralmente em parsecs (pc). Vamos então usar estes dados para na prática ver como funciona a estimativa da constante de Hubble.

## Fazendo a regressão linear

No nosso modelo vamos considerar que a variável, $v^{*} = v-v_{s}$, será explicada pela distância $\mathbf{r}$ e não vamos considerar que existe um valor de $v^{*}$ quando $\mathbf{r}=0$, isto é, consideramos que $b=0$. Temos uma razão pra isso: a expansão é um efeito observável para grandes distâncias. Não conseguimos medir a taxa de expansão do universo dentro da nossa própria galáxia.  Usando os dados da tabela que está contida no artigo do Hubble:

```python
import pandas as pd
dados_Hubble = pd.read_csv("HubbleDados.csv")
```

```python
dados_Hubble = dados_Hubble.dropna()
```

Escrevendo a variável $v^{*}$ e selecionando as distâncias:

```python
v_estrela = dados_Hubble[r"$v$"] - dados_Hubble[r"$v_{s}$"] 
distancia = dados_Hubble[r"$r$"]
```

Vamos ver se é possível notar visualmente alguma relação linear entre as variáveis:


```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(9,5))
plt.grid(True, linestyle='--', alpha=0.25)
plt.scatter(dados_Hubble[r"$r$"], dados_Hubble[r"$v$"], color='#7D3C98')
plt.xlabel(r"$r$ (Mpc)", fontsize=15)
plt.ylabel(r" $v$"+"\n"+  "(km/s)", rotation=0, labelpad=29, fontsize=15 )
plt.title("Gráfico de dispersão da Amostra")
```

Podemos perceber que uma relação linear pode sim estar presente entre essas variáveis. Vamos lembrar que além dá aparente relação entre as duas existem alguns critérios que essas variáveis devem seguir, embora não abordamos estes critérios como a homocesdasticidade (que é basicamente exigir que a variância dos erros será constante) partindo para a regressão, usando a expressão para o coeficente $a$, exrevemos:

```python
H0 = (v_estrela.dot(distancia) - v_estrela.mean()*distancia.sum() )/\
(distancia.dot(distancia) - distancia.mean()*distancia.sum())
```

E escrevemos a equação da reta de regressão:

```python
v_chapeu = H0*distancia
```

Vejamos como ficou nosso resultado:

```python
fig = plt.figure(figsize=(9,5))
plt.grid(True, linestyle='--', alpha=0.25)
plt.scatter(dados_Hubble[r"$r$"], dados_Hubble[r"$v$"], color='#7D3C98')
plt.plot(distancia,v_chapeu, color='#52BE80', linestyle="dotted", lw=2, label=r"$H_{0}=505$ Km/s/Mpc")
plt.legend(loc="upper left")

plt.xlabel(r"$r$ (Mpc)", fontsize=15)
plt.ylabel(r" $v$"+"\n"+  "(km/s)", rotation=0, labelpad=29, fontsize=15 )
plt.title("Gráfico de dispersão da Amostra")
```

Podemos ver que a estimativa para $H_{0}$, isto é o coeficiente angular da reta foi para todos os efeitos igual ao valor encontrado pelo Hubble em 1929. 

$$
H_{0} = 500 \,\,\, \text{km}/\text{s}/\text{Mpc}
$$

```python
print(round(H0,2))
```

Atualmente além das cefeidas, outras velas padrão como supernovas do tipo IA nos ajudam a medir a taxa de expansão do universo. O melhor valor medido hoje para $H_{0}$ está em torno de $70$ Km/s/Mpc. Poderíamos nos perguntar: quão bom é o desempenho desse modelo para explicar as observações? Para responder a essa pergunta podemos calcular o R$^{2}$, e ter uma ideia quantitativa do desempenho dessa relação. 

```python
s_e = v_estrela - v_chapeu
s_t = v_estrela - v_estrela.mean()

r2 = 1 - s_e.dot(s_e)/s_t.dot(s_t)
```

```python
print(r"O coeficiente de determinação $R^{2}$ do modelo é: ", r2)
```

É um ótimo valor para o R$^{2}$, você não vai ver muitos modelos reais com um coeficiente de determinação tão bom assim.
